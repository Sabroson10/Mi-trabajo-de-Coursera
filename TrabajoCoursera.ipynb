```python
import yfinance as yf

# Descargar los datos financieros de Tesla
tesla = yf.Ticker("TSLA")

# Extraer el historial de ingresos
tesla_revenue = tesla.financials.loc['Total Revenue'].T  # Transponer para poner los años en columnas

# Mostrar las últimas cinco filas
print(tesla_revenue.tail())
```

```python
import yfinance as yf

# Descargar los datos financieros de GameStop
gme = yf.Ticker("GME")

# Extraer el historial de ingresos (financials)
gme_revenue = gme.financials.loc['Total Revenue'].T  # Transponer para mostrar los ingresos por año

# Mostrar las últimas cinco filas
print(gme_revenue.tail())

# Guardar los datos en un archivo CSV
gme_revenue.to_csv('gme_revenue.csv', index=False)
```

```python
!pip install matplotlib
import yfinance as yf
import matplotlib.pyplot as plt

# Función para crear el gráfico
def make_graph(data, title):
    plt.figure(figsize=(10,6))
    plt.plot(data['Date'], data['Close'], label='Tesla Stock Price', color='blue')
    plt.title(title)
    plt.xlabel('Date')
    plt.ylabel('Close Price (USD)')
    plt.grid(True)
    plt.legend()
    plt.show()

# Descargar los datos de Tesla
tesla_data = yf.download('TSLA', start='2020-01-01', end='2023-01-01')

# Restablecer el índice para que la columna 'Date' esté disponible
tesla_data_reset = tesla_data.reset_index()

# Llamar a la función make_graph para graficar los datos
make_graph(tesla_data_reset, 'Tesla Stock Price (2020-2023)')
```
